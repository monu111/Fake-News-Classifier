# LSTM(Long Short term memory)

### it is type of RNN architecture. which basically use for sequential info. it can add or remove the previous info. with recent info. to perform persent task.


- it reslove the Gradient vanishing probem & Long term of dependency.
- the Sequential info. regulated by the gates. 
#### there are 4 component in LSTM :

__1. cell state__: it can keep or forget the info. it work like memeory. where we can keep the sequential info. and if we want to forget the previous  info. it is  regulated by forget gate.

__2. Forget gate :__ where we have sigmoid activation function. To remember or forget the info. if sigmoid o/p 1 that's means we can keep the info. if 0 that's means we have to forget the info. it connected with __cell state__.


__3. input gate :__ when new candidate info. come then we can feed to the input gate. which can control how much new  info we can keep or forget. then after the value will go to __cell state__. now cell state is updated.

__4. output gate :__ Now we have to decide which value we are going to o/p. we feed  updated cell state value to output gate. then it will give the final o/p.

__Ad:__ Most of the NLP usecase it perform well with better accuracy. Slower compare to Normal RNN.

__Dis:__   Prone to the  overfiiting.




# GRU( Gated recurrent Unit)

#### GRU working are simliar to the LSTM. but there are only 2 gate & there is no Cell state. 

__Update gate :__ update gate similare to the i/p gate and Forget gate of LSTM. It decide what info. to throw away & what new info. to add __or__ The update gate helps the model to determine how much of the past information (from previous time steps) needs to be passed along to the future.
 
__Reset gate :__ it decide how much past info. should forget.
 
 

__ad:__ train the data fastly compare to LSTM. optimize quciky compare to LSTM.
